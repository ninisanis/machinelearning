# 01. Dasar Machine Learning

### A. Tujuan Pembelajaran

Setelah mempelajari bab ini diharapkan dapat memfasilitasi pemahaman mahasiswa tentang konsep dasar machine learning. Mahasiswa diharapkan dapat menjelaskan definisi dan aplikasi dari machine learning, memahami berbagai jenis machine learning seperti supervised learning, unsupervised learning, dan pembelajaran terbimbing, pembelajaran tidak terbimbing, dan semi-supervised learning. Selain itu, mahasiswa juga harus mampu menjelaskan perbedaan antara jenis-jenis machine learning tersebut serta pemahaman tentang bagaimana algoritma machine learning bekerja.

Dengan penguasaan mendalam terhadap konsep-konsep tersebut, diharapkan mahasiswa dapat memulai tahap eksplorasi dalam memahami dan mengaplikasikan metode-metode machine learning. Penerapan ini tidak terbatas pada lingkungan akademik semata, tetapi juga pada beragam situasi dan tantangan yang ada dalam konteks dunia nyata. 


### B. Apa itu Machine Learning ?

Komputer, melalui pendekatan machine learning, memiliki kapasitas untuk “belajar” dari kumpulan data yang diberikan, yang analog dengan proses belajar yang dialami oleh manusia. Dalam konteks manusia, proses belajar melibatkan akuisisi pengetahuan dari berbagai sumber seperti guru, orang tua, literatur, media visual, dan interaksi dengan lingkungan. Cara belajar ini juga diaplikasikan ke dalam komputer untuk meningkatkan kecerdasannya. Namun, untuk memfasilitasi proses ini, penting untuk memberikan komputer instruksi mengenai bagaimana cara belajar dari data yang diberikan, suatu konsep yang akan dibahas secara mendalam dalam bagian metode dan algoritma.

Pada tahap ini, machine learning mengandalkan serangkaian algoritma yang secara berulang-ulang membimbing komputer untuk belajar dari data. Proses ini melibatkan penyediaan input berupa data yang berkelanjutan kepada komputer. Paralel dengan ini, algoritma beradaptasi dan bertransformasi secara dinamis seiring dengan data yang diterima. Akibatnya, peningkatan signifikan dapat diamati dalam kapabilitas komputer dalam pemrosesan data, pemahaman, dan bahkan dalam memprediksi hasil atau “outcome”.

Jadi dapat disimpulkan definisi dari machine learning adalah suatu disiplin dalam ilmu komputer yang memanfaatkan metode statistika, sehingga komputer diberikan kemampuan untuk “belajar” dari data tanpa perlu pengaturan atau pemrograman yang eksplisit. 


### C. Sejarah Singkat Machine Learning

Machine learning sebagai suatu disiplin ilmu memiliki sejarah yang panjang dan menarik. Awal mula machine learning tidak lepas dari perkembangan ilmu komputer dan kecerdasan buatan. Dalam konteks sejarah, evolusi machine learning dapat dilihat melalui berbagai fase perkembangan teknologi yang berbeda (Russell and Norvig, 2009).

Pada tahun 1950-an, Alan Turing, yang dianggap sebagai bapak komputasi modern, memperkenalkan konsep “mesin belajar”, yang kemudian menjadi dasar dari perkembangan machine learning. Dalam artikelnya yang berjudul "Computing Machinery and Intelligence", Turing merumuskan apa yang sekarang dikenal sebagai Tes Turing, suatu metode untuk menentukan apakah mesin bisa “berpikir”.
Kemudian, pada tahun 1959, Arthur Samuel mengembangkan program catur yang mampu belajar dari pengalamannya sendiri, yang merupakan aplikasi nyata pertama dari machine learning. Pada saat yang sama, diadakan konferensi Dartmouth yang memfokuskan pada “belajar otomatis”, di mana istilah kecerdasan buatan pertama kali diperkenalkan.

Perkembangan selanjutnya dalam bidang machine learning terjadi pada tahun 1980-an dan 1990-an dengan kemunculan berbagai algoritma baru seperti decision trees, support vector machines, dan neural networks. Pada era ini, terjadi peningkatan signifikan dalam penggunaan machine learning, terutama dalam bidang pengenalan pola dan sistem rekomendasi.

Dalam era modern, machine learning telah menjadi kunci dalam berbagai aplikasi teknologi tinggi, seperti pencarian web, deteksi spam email, dan pengenalan suara. Terobosan seperti “deep learning” telah membuka jalan untuk kemajuan luar biasa dalam bidang ini. Dengan demikian, sejarah machine learning adalah cerita tentang bagaimana komputer semakin mendekati kemampuan manusia dalam belajar dan membuat keputusan.

Mengulas sejarah perkembangan machine learning ini diharapkan dapat membuka pemahaman kita bahwa masih banyak tantangan yang menanti untuk dipecahkan. Anda mungkin akan menjadi bagian dari masa depan dalam sejarah machine learning.


### D. Penggunaan Machine Learning

Penggunaan machine learning telah merambah hampir semua aspek kehidupan kita sehari-hari. Salah satu penggunaan utama dari machine learning yang paling jelas adalah dalam bidang teknologi pencarian. Mesin pencari atau search engine seperti Google menggunakan algoritma machine learning untuk meningkatkan relevansi dan kualitas hasil pencarian mereka. Selain itu, penggunaan machine learning juga sangat penting dalam teknologi pengenalan suara, yang telah banyak digunakan dalam asisten virtual seperti Siri dan Alexa.

Di bidang kesehatan, machine learning juga memainkan peran penting. Dengan menggabungkan machine learning dan big data, para peneliti dapat menganalisis dan memahami pola dalam data kesehatan yang sangat besar dan kompleks, yang dapat digunakan untuk mendeteksi penyakit lebih awal atau meramalkan perkembangan penyakit. Selain itu, machine learning juga digunakan dalam pengembangan obat baru dan peningkatan hasil diagnosis.

Dalam bidang bisnis dan ekonomi, machine learning juga digunakan secara luas. Sistem rekomendasi, seperti yang digunakan oleh Amazon dan Netflix, menggunakan machine learning untuk memahami preferensi pelanggan dan memberikan rekomendasi produk atau konten yang lebih personal. Selain itu, machine learning juga digunakan dalam analisis sentimen untuk memahami opini dan sentimen konsumen terhadap produk atau layanan.

Tak kalah pentingnya, penerapan machine learning dalam bidang keamanan siber atau cyber security menjadi amat krusial. Menyadari bahwa serangan siber menjadi ancaman yang semakin serius dalam era digital ini, penggunaan machine learning mampu membantu dalam deteksi ancaman dan respon yang lebih cepat dan akurat. Algoritma machine learning dapat dilatih untuk mengidentifikasi pola-pola dalam data trafik jaringan yang dapat menunjukkan adanya aktivitas yang mencurigakan atau tidak biasa, yang kemungkinan merupakan serangan siber. Selain itu, teknologi machine learning juga digunakan dalam sistem otentikasi dan verifikasi pengguna, dimana algoritma dapat belajar untuk mengenali pola-pola perilaku pengguna dan mendeteksi jika ada aktivitas yang tidak sesuai dengan pola yang biasa dilakukan oleh pengguna, sehingga menandakan kemungkinan adanya penyalahgunaan akun atau data.

Machine learning juga digunakan dalam bidang pendidikan, di mana digunakan untuk mengembangkan sistem pembelajaran adaptif yang dapat menyesuaikan materi dan metode pengajaran berdasarkan kebutuhan dan kemampuan belajar individu. Selain itu, machine learning juga digunakan dalam penelitian pendidikan untuk memahami pola dan tren dalam data pendidikan.

Terakhir, penggunaan machine learning dalam bidang lingkungan juga semakin meningkat. Misalnya, dalam upaya mitigasi perubahan iklim, machine learning digunakan untuk memprediksi dan meramalkan perubahan iklim serta dampaknya. Juga digunakan dalam konservasi keanekaragaman hayati, di mana machine learning digunakan untuk memahami pola dan tren dalam data keanekaragaman hayati dan membantu dalam pengambilan keputusan konservasi.


### E. Proses Machine Learning

Proses machine learning adalah serangkaian langkah yang diambil untuk membangun dan melatih model dari data. Sebelum kita mempelajari beberapa tipe atau jenis machine learning, ada baiknya kita memahami terlebih dahulu beberapa tahapan utama dalam proses machine learning. Pemahaman yang baik tentang setiap tahap ini sangat penting untuk pembuatan model yang efektif (Brownlee, 2021).

Pada tahap pertama, yang disebut dengan pemahaman masalah, kita perlu memahami tujuan yang ingin dicapai, dan bagaimana machine learning dapat membantu mencapai tujuan tersebut. Setelah memahami permasalahan yang ingin dipecahkan dengan machine learning, maka kita akan mengumpulkan data dari berbagai sumber yang ada. Selanjutnya sangat penting juga untuk memahami secara mendalam tentang data yang akan digunakan. Analisis eksploratif data, yaitu proses mengidentifikasi pola dan tren dalam data, sering kali dilakukan pada tahap ini.

Tahap berikutnya adalah pra-pemrosesan data. Pada tahap ini, data dibersihkan dan diubah menjadi format yang dapat digunakan oleh algoritma machine learning. Ini bisa termasuk penanganan data yang hilang, penghapusan noise, dan normalisasi data. Teknik pemilihan fitur juga sering digunakan pada tahap ini untuk memilih fitur yang paling relevan untuk model.

Setelah data siap, tahap berikutnya adalah memilih dan melatih model. Pada tahapan inilah kita akan menerapkan metode atau algoritma machine learning. Ada berbagai algoritma machine learning yang dapat digunakan, dan pilihan algoritma ini sangat bergantung pada masalah yang dihadapi dan data yang ada. Setelah model dipilih, data latihan digunakan untuk melatih model tersebut.

Tahap keempat adalah evaluasi model. Pada tahap ini, model diuji dengan menggunakan data tes untuk melihat sejauh mana model tersebut dapat memprediksi hasil dengan akurat. Metrik evaluasi seperti akurasi, presisi, dan recall sering digunakan pada tahap ini.


### F. Jenis Machine Learning

Terdapat tiga jenis utama algoritma dalam machine learning menurut para pakar machine learning, yakni pembelajaran terbimbing (supervised learning), pembelajaran tak terbimbing (unsupervised learning), dan pembelajaran semi-terbimbing (semi-supervised learning). Pada supervised learning, kita membuat model dengan data latih yang sudah kita tahu input dan outputnya. Layaknya seperti guru yang memberikan soal dan jawabannya ke siswa, lalu siswa belajar dari soal dan jawaban tersebut. Sedangkan dalam unsupervised learning, kita hanya memiliki input tetapi tidak ada jawaban atau output yang jelas. Ini seperti siswa yang diberikan soal tetapi tidak diberikan jawabannya, sehingga siswa harus mencari jawabannya sendiri. Sementara pada semi-supervised learning, model dibuat dengan campuran data yang ada jawabannya (berlabel) dan data yang tidak ada jawabannya (tidak berlabel). Itu seperti seorang siswa yang belajar dari beberapa soal yang ada jawabannya dan beberapa soal lainnya yang tidak ada jawabannya.

Selain ketiga jenis algoritma tersebut, ada pakar yang menambahkan yakni Reinforcement Learning. Berbeda dengan jenis algoritma yang sebelumnya, reinforcement learning (RL) adalah sebuah paradigma dalam machine learning yang berorientasi pada bagaimana sebuah agen dapat mempelajari strategi optimal dalam suatu lingkungan berdasarkan interaksi dan feedback yang diterimanya, yang umumnya dalam bentuk reward atau penghargaan. Dalam RL, agen mencoba untuk memaksimalkan total reward yang diterima melalui serangkaian tindakan atau keputusan yang diambil. Konsep dasar RL sangat erat dengan prinsip belajar melalui pengalaman dalam psikologi, yang menyatakan bahwa suatu entitas dapat belajar dan mengubah perilakunya berdasarkan konsekuensi dari perilaku tersebut, berupa reward (penguatan positif) atau punishment (penguatan negatif). Dalam konteks RL, agen belajar strategi atau kebijakan yang optimal berdasarkan pengalaman yang diperolehnya dalam interaksi dengan lingkungan. Sebuah contoh sederhana dari penerapan Reinforcement Learning (RL) adalah proses belajar bermain game catur. Bayangkan Anda adalah sebuah agen (dalam hal ini, pemain catur) yang belum pernah bermain catur sebelumnya. Anda hanya diberi aturan dasar tentang bagaimana setiap bidak dapat bergerak, tetapi tidak diberi strategi atau taktik apa pun tentang bagaimana memenangkan permainan. 

Jika kita lihat, machine learning memiliki kesamaan dengan data mining. Data mining dan machine learning saling berkaitan dan sering kali digunakan secara bersamaan dalam bidang ilmu data. Data mining adalah proses mencari pola dan keterkaitan dalam data yang besar dan kompleks, sementara machine learning adalah teknik yang digunakan untuk 'mengajar' komputer bagaimana cara belajar dari data dan membuat keputusan atau prediksi berdasarkan pola tersebut. Dalam banyak kasus, algoritma machine learning digunakan sebagai bagian dari proses data mining. Misalnya, algoritma klasifikasi atau klastering dalam machine learning dapat digunakan untuk menemukan pola atau struktur dalam data dalam proses data mining. Dengan kata lain, machine learning adalah salah satu teknik utama yang digunakan dalam data mining untuk memahami dan mengekstrak informasi dari data. Data mining lebih berfokus pada penemuan pengetahuan dan wawasan baru dari data, machine learning lebih berfokus pada pembuatan model prediktif dan deskriptif yang dapat belajar dan mengadaptasi pola dari data.. Oleh karena itu, hubungan antara data mining dan machine learning adalah saling melengkapi dan membantu dalam proses penemuan pengetahuan dari data.
Sehingga dapat dikatakan bahwa peran dalam data mining, juga dapat dilakukan di machine learning, seperti:
1.	Estimasi (Supervised Learning).
Estimasi merujuk pada proses memprediksi nilai kontinu atau numerik. Estimasi dilakukan dengan menggunakan variabel input untuk memprediksi nilai variabel output. Misalnya, menggunakan data demografis konsumen untuk memprediksi pendapatan mereka.
2.	Forecasting (Supervised Learning)
Forecasting adalah proses memprediksi tren atau pergerakan data di masa depan berdasarkan pola historis. Misalnya, memprediksi penjualan produk pada kuartal berikutnya berdasarkan data penjualan sebelumnya.
3.	Klasifikasi (Supervised Learning)
Klasifikasi adalah proses mengkategorikan data ke dalam kelas atau grup berdasarkan fitur-fitur data tersebut. Misalnya, memprediksi apakah seorang pelanggan akan melakukan pembelian atau tidak berdasarkan data belanja dan perilaku mereka.
4.	Klastering (Unsupervised Learning)
Klastering adalah proses mengelompokkan data ke dalam kluster atau grup berdasarkan kemiripan data tersebut, tanpa perlu label kelas sebelumnya. Misalnya, mengelompokkan pelanggan ke dalam segmen berdasarkan perilaku belanja mereka.
5.	Asosiasi (Unsupervised Learning)
Asosiasi adalah proses mencari pola hubungan antara item atau variabel dalam data. Misalnya, mencari pola barang yang sering dibeli bersama dalam transaksi belanja.

Jadi pada dasarnya, algoritma dari machine learning akan berkaitan dengan 5 (lima) hal tersebut. Pada pembahasan di bab selanjutnya kita akan membahas secara detail satu-persatu jenis algoritma machine learning.


### G. Machine Learning vs AI

Kecerdasan buatan (Artificial Intelligence/AI) dan machine learning adalah dua konsep yang sering menjadi bahan diskusi dalam bidang teknologi informasi, dan kerap menjadi topik yang membingungkan. AI adalah suatu disiplin ilmu yang berupaya menciptakan sistem atau mesin yang mampu meniru kecerdasan manusia. Sementara itu, machine learning adalah cabang dari AI yang mengkhususkan diri dalam pengembangan algoritma dan model yang memberikan kemampuan belajar pada mesin berdasarkan data (Russell and Norvig, 2009). Sebagai bagian dari AI, machine learning turut menyumbang dalam pencapaian tujuan AI secara keseluruhan.

Dalam bidang AI, ada dua klasifikasi utama yakni AI lemah dan AI kuat. AI lemah mencakup sistem yang dibuat dan dilatih untuk tugas-tugas spesifik, misalnya asisten virtual atau sistem rekomendasi. Sementara itu, AI kuat adalah konsep yang mengacu pada sistem yang memiliki kemampuan kognitif secara umum, serupa dengan manusia. Hingga saat ini, AI kuat masih berada dalam tahap penelitian dan kebanyakan aplikasi AI yang ada merupakan AI lemah.

Machine learning, di sisi lain, berfokus pada proses belajar mesin melalui analisis data. Algoritma machine learning menggunakan prinsip-prinsip statistik dan matematika untuk mengeksplorasi dan mencari pola dalam data, yang kemudian menjadi dasar mesin untuk belajar dan membuat prediksi atau keputusan.
Sebagai bagian dari machine learning, ada juga konsep deep learning. Deep learning memanfaatkan jaringan saraf tiruan dengan beberapa lapisan (sering disebut sebagai "deep" networks) untuk melaksanakan tugas-tugas yang lebih rumit. Oleh karena itu, deep learning merupakan sub-bidang dari machine learning, dan machine learning sendiri merupakan bagian dari AI.

Dalam praktiknya, AI dan machine learning saling terkait dan berkontribusi dalam perkembangan teknologi yang kita lihat saat ini. Istilah AI dan machine learning sering digunakan secara bergantian, meskipun pada dasarnya mereka memiliki arti yang berbeda. Dengan memahami hubungan dan perbedaan antara AI dan machine learning, kita dapat lebih memahami bagaimana teknologi ini dapat digunakan untuk membantu memecahkan tantangan dalam berbagai sektor di dunia nyata.


### H. Alat dan Bahan

Untuk memulai pembelajaran machine learning, dua aspek penting yang harus dipersiapkan adalah perangkat keras (hardware) dan perangkat lunak (software). Dari segi perangkat keras, komputer dengan spesifikasi yang mumpuni sangat dibutuhkan. Mengingat machine learning melibatkan pengolahan data dalam jumlah besar dan rumit, komputer dengan prosesor yang kuat, memori yang cukup, serta kartu grafis (GPU) berkualitas tinggi menjadi sangat penting. GPU, khususnya, sangat berperan dalam proses pembelajaran deep learning, di mana perhitungan matematis paralel dapat dilakukan dengan lebih efisien dibandingkan prosesor biasa (CPU).

Sementara itu, perangkat lunak untuk belajar machine learning meliputi sistem operasi, bahasa pemrograman, dan pustaka atau library khusus machine learning. Python dan R menjadi bahasa pemrograman yang paling populer dalam bidang ini karena kedua bahasa ini menyediakan banyak pustaka yang mendukung implementasi algoritma machine learning. Beberapa pustaka yang populer di Python antara lain adalah Scikit-learn, TensorFlow, Keras, dan PyTorch, sementara di R populer pustaka seperti Caret dan MLR. Selain itu, sistem operasi yang paling umum digunakan adalah Linux karena stabilitas dan fleksibilitasnya dalam melakukan komputasi ilmiah, meskipun Windows dan macOS juga bisa digunakan.

Pada buku ini, akan menggunakan  Sistem Operasi Windows dan bahasa pemrograman Python. Sehingga perlu menginstall beberapa hal yang dibutuhkan. seperti Anaconda, sebuah platform ilmu data yang populer dan membantu dalam menginstal dan mengelola paket Python dan R, yang banyak digunakan dalam analisis data dan machine learning. Anaconda membawa konsol Python (IPython), Jupyter Notebook, dan Spyder (IDE Python). Paket-paket ini membantu dalam mempersiapkan lingkungan kerja yang optimal untuk machine learning. IPython dan Jupyter Notebook digunakan untuk interaksi langsung dengan kode, yang memungkinkan pengecekan dan pengujian algoritma machine learning. Spyder, di sisi lain, adalah lingkungan pengembangan yang terintegrasi yang sangat berguna untuk penulisan dan pengujian kode dalam skala besar.
