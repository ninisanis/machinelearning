```python
import pandas as ps

df = pd.read_csv('WHR_2023.csv')
df
```

```python
size_mapping = {'Western Europe': 1,
               'Middle East and North Africa': 2,
               'North America and ANZ': 3,
               'Central and Eastern Europe' :4 ,
               'Latin America and Caribbean': 5,
               'Southeast Asia': 6,
               'East Asia': 7,
               'Commonwealth of Independent States': 8,
               'Sub-Saharan Africa': 9,
               'South Asia': 10,
                'Southeast Asia': 11}

df['region'] = df['region'].map(size_mapping)
df
```

```python
inv_size_mapping = {v: k for k, v in size_mapping.items()}
df['region'].map(inv_size_mapping)
```

```python
import numpy as np

#create a mapping dict
#to convert class labels from strings to integers
class_mapping = {label: idx for idx, label in enumerate(np.unique(df['country']))}
class_mapping
```

```python
#to convert class labels from strings to integers
df['country'] = df['country'].map(class_mapping)
df
```

```python
#reverse the class labels maping

inv_class_mapping = {v: k for k, v in class_mapping.items()}
df['country'] = df['country'].map(inv_class_mapping)
df
```

```python
from sklearn.preprocessing import LabelEncoder

#label encoding with sklearn;s labelEncoder
class_le = LabelEncoder()
y = class_le.fit_transform(df['country'].values)
y
```

```python
#reverse mapping
class_le.inverse_transform(y)
```

```python
X = df[['country', 'happiness_score', 'region',]].values
color_le = LabelEncoder()
X[:, 0] = color_le.fit_transform(X[:, 0])
X
```

```python
from sklearn.preprocessing import OneHotEncoder
X = df[['social_support', 'freedom_to_make_life_choices', 'perceptions_of_corruption']].values
color_one = OneHotEncoder()
color_one.fit_transform(X[:, 0].reshape(-1, 1)).toarray()
```

```python
from sklearn.compose import ColumnTransformer

X = df[['social_support', 'freedom_to_make_life_choices', 'perceptions_of_corruption']].values
c_transf = ColumnTransformer([('onehot', OneHotEncoder(), [0]),
                             ('nothing', 'passthrough', [1, 2])])
c_transf.fit_transform(X).astype(float).toarray()
```

```python
#one-hot encoding via pandas
pd.get_dummies(df[['region', 'happiness_score', 'freedom_to_make_life_choices']])
```

```python
#multicollinearity guard in get_dummies
pd.get_dummies(df[['region', 'happiness_score', 'freedom_to_make_life_choices']], drop_first=True)
```

```python
#multicollinearity guard for the OneHotEncoder
color_one = OneHotEncoder(categories='auto', drop='first')
c_transf = ColumnTransformer([('onehot', color_one, [0]),
                             ('nothing', 'passthrough', [1, 2])])
c_transf.fit_transform(X).astype(float).toarray()
```

```python
import pandas as pd
import numpy as np

# Membaca dataset wine dari URL
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
df_wine = pd.read_csv(url, header=None)

# Menyertakan nama kolom
column_names = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                'Alcalinity of ash', 'Magnesium', 'Total phenols',
                'Flavanoids', 'Nonflavonoid phenols', 'Proanthocyanins',
                'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']

df_wine.columns = column_names

# Menampilkan unique class labels
print('Class labels', np.unique(df_wine['Class label']))

df_wine.head()
```

```python
from sklearn.model_selection import train_test_split

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test =\
train_test_split(X,y,
                test_size = 0.3,
                random_state = 0,
                stratify = y)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
ex = np.array([0, 1, 2, 3, 4, 5])

print('standardized: ', (ex - ex.mean()) / ex.std())

#normalized
print('normalized: ', (ex - ex.min()) / (ex.max() - ex.min()))
```
